# 逻辑回归


## Sigmoid 函数



```python
import matplotlib.pyplot as plt
import numpy as np

# 使用黑体
plt.rcParams['font.family'] = ['SimHei']
plt.rcParams['axes.unicode_minus'] = False

z = np.linspace(-6, 6)
g = 1/(1+np.exp(-z))

plt.figure(figsize=(8, 4))  # 创建画布
plt.title('Sigmoid 函数')

ax = plt.gca()      # get current axis 获得坐标轴对象
# 将右边 上边的两条边颜色设置为空 其实就相当于抹掉这两条边
ax.spines['right'].set_color('none') 
ax.spines['top'].set_color('none')         
# 指定下边的边作为 x 轴   指定左边的边为 y 轴
ax.xaxis.set_ticks_position('bottom')   
ax.yaxis.set_ticks_position('left')          
#指定 data 设置的bottom(也就是指定的x轴)绑定到y轴的0这个点上
ax.spines['bottom'].set_position(('data', 0))   
ax.spines['left'].set_position(('data', 0))

# 在轴上添加一条水平线, 水平坐标为 1, 颜色为黑, 线条样式为虚, 线宽为 1 磅
plt.axhline(1, color='black', linestyle='--', linewidth=1)

plt.plot(z, g, label=r"$\sigma(z) = \frac{1}{1 + e^{-z}}$")
plt.xlabel("z")
plt.legend(fontsize=10, loc='center right')
```
